```python
def load_and_check_dataset(language: str = "eng",
                           path: str = "BabyLM-community/babylm-eng"):
    ds = load_dataset(path, token=token)
    print(f"Dataset language is: {language} and path is: {path}")
    print(f"Dataset format is: {ds}")
    print("--------------------------------\nExample sample from the dataset:")
    print(ds['train'][1])
    num_tokens = 0
    for split in ds.keys():
        for sample in ds[split]:
            num_tokens += sample['num-tokens']
    print(f"Total number of tokens: {num_tokens}\n")
    print("=" * 40)
    return ds
```

```python
def save_text(ds, path):
    with open(path, "w") as f:
        for sample in ds:
            text = sample["text"].replace("\n", " ")
            f.write(text.strip() + "\n")
```

```python
save_text(ds_eng['train'], "data/eng.txt")
save_text(ds_deu['train'], "data/deu.txt")
save_text(ds_dut['train'], "data/nld.txt")
```
